## Simple Linear Regression

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
X = np.array([ 0.387, 0.723, 1.00, 1.52, 5.20, 9.54, 19.2, 30.1, 39.5 ])
Y = np.array([ 0.241, 0.615, 1.00, 1.88, 11.9, 29.5, 84.0, 165.0, 248 ])
N = 9
```

```python
Ys = np.log(Y)
Xs = np.log(X)
```

```python
np.corrcoef(Ys, Xs)
```

```python
sns.residplot(x=Xs, y=Ys, lowess=True, color="g")
```

```python
df = pd.DataFrame({'Xs': Xs,
                   'Ys': Ys})
                  
```

```python
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols('Ys ~ Xs', data=df).fit()
print(model.summary())
```

```python
fig = plt.figure(figsize=(12,8))

#produce regression plots
fig = sm.graphics.plot_regress_exog(model, 'Xs', fig=fig)
```

### QQ-plot

```python
import statsmodels.api as sm
sm.qqplot(Xs, line='s')
plt.title("X distribution")
plt.show()
```

```python
sm.qqplot(Ys, line='s')
plt.title("Y distribution")
plt.show()
```
