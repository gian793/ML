## Gaussian Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics.pairwise import rbf_kernel
```

```python
# This is the true unknown function we are trying to approximate
f = lambda x: np.sin(0.9*x).flatten()
```

```python
def plot_f(x,y, title, symbol, color):
    # PLOTS:
    plt.figure(1, figsize=(10, 5))
    plt.clf()
    plt.plot(x, y, symbol, color = color)
    plt.title(title)
    plt.axis([-5, 5, -3, 3])
```

```python
# Example 
Xe = np.linspace(-5,5, 51)
Ye = f(Xe)

plot_f(Xe, Ye, 'Example: function to predict', symbol='+', color='r')
```

#### Train points

```python
N = 10         # number of training points.
n = 50         # number of test points.
s = 0.00005    # noise variance.

# Sample some input points 
X = np.random.uniform(-5, 5, size=(N,1))
print('X:')
print(X)

print('f(X):')
print(f(X))

# Add noise
y = f(X) + s*np.random.randn(N)
print('y = f(X) + noise: ')
print(y)

kernelParameter = 0.1
Gamma= 0.5/kernelParameter
    
K = rbf_kernel(X,X, Gamma)

L = np.linalg.cholesky(K + s*np.eye(N))

plot_f(X,y, 'Training points', symbol='+', color='b')
```

```python
# We pick up n points equally spaced on x-axis in the range (-5, 5)
x = np.linspace(-5,5, n).reshape(-1,1)

y = np.zeros(n)

plt.figure(1, figsize=(16, 5))

plt.clf()
plt.plot(x, y, marker='x', color = 'b')

plt.axis([-5, 5, -3, 3])
```

```python
# We sample 3 times n random points from a standard normal 
sample_f = np.random.normal(size=(n,3))

plt.figure(1, figsize=(16, 5))

plt.clf()
plt.plot(x, sample_f[:,0], marker='o', color = 'r')
plt.plot(x, sample_f[:,1], marker='o', color = 'b')
plt.plot(x, sample_f[:,2], marker='o', color = 'c')

plt.axis([-5, 5, -3, 3])
```

```python
sample_f
```

```python
# We calculate the covariance matrix (kernel) on the 50 equally spaced points 
Ks = rbf_kernel(x,x, Gamma)

# Cholesky matrix calculation
Ls = np.linalg.cholesky(Ks + s*np.eye(n))

# Note that Cholesky matrix is used to create correlations among random variables. 
# Suppose that x1 and x2 are independent standard normal variables. 
# The matrix Ls can be used to create new variables z1 and z2 such that 
# the covariance of z1 and z2 equals Σ
# In particular, if you generate p standard normal variates, the Cholesky transformation maps the variables into variables
# for the multivariate normal distribution with covariance matrix Σ and centered at the origin (denoted MVN(0, Σ)).

# The 3 prior functions are created by correlating the randomly sampled points 
f_prior = np.dot(Ls, sample_f)

plt.figure(1, figsize=(16, 5))
plt.clf()
plt.plot(x, f_prior[:,0], marker='o', color = 'r')
plt.plot(x, f_prior[:,1], marker='o', color = 'b')
plt.plot(x, f_prior[:,2], marker='o', color = 'c')

plt.axis([-5, 5, -3, 3])
```

```python
# The resulting priors have zero-mean (this is result clearer by increasing n)

print(f_prior[:,0].mean())
print(f_prior[:,1].mean())
print(f_prior[:,2].mean())
```

#### Test points

```python
# Test points we're going to make predictions at
Xtest = np.linspace(-5, 5, n).reshape(-1,1)

z = np.zeros(n)
plot_f(Xtest, z, 'Test points', symbol='g+', color='b')

# Mean at our test points.
Lk = np.linalg.solve(L, rbf_kernel(X, Xtest, Gamma))
mu = np.dot(Lk.T, np.linalg.solve(L, y))

print(mu)

# Variance at our test points.
K_ = rbf_kernel(Xtest, Xtest, Gamma)
s2 = np.diag(K_) - np.sum(Lk**2, axis=0)
s = np.sqrt(s2)


# PLOTS:
plt.figure(1)
plt.clf()
plt.plot(X, y, 'ro', ms=10)
plt.plot(Xtest, f(Xtest), 'b-')
plt.gca().fill_between(Xtest.flat, mu-3*s, mu+3*s, color="#dddddd")
plt.plot(Xtest, mu, 'r+', lw=2)
plt.savefig('predictive.png', bbox_inches='tight')
plt.title('Mean predictions plus 3 st.deviations')
plt.axis([-5, 5, -3, 3])
```
